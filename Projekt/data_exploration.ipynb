````python
popularity_bins = [0, 20, 40, 60, 80, 100]

df['popularity_labels'] = pd.cut(df['popularity'], bins=popularity_bins, labels=['Bardzo niska', 'Niska', 'Srednia', 'Wysoka', 'Bardzo wysoka'])
df['popularity_labels'] = np.where(df['popularity'] == 0, 'Bardzo niska', df['popularity_labels'])


df
````

````python
df_labeled = df.drop('popularity', axis=1)
````

````python
df_labeled['popularity_labels_encoded'] = df_labeled['popularity_labels'].replace(
    { 'Bardzo niska': 0, 'Niska': 1, 'Srednia': 2, 'Wysoka': 3, 'Bardzo wysoka': 4}
)

df_labeled
````

````python
df_labeled["popularity_labels_encoded"].unique()
````

````python
df_labeled['popularity_labels'].unique()
````

````python
df_labeled.drop('popularity_labels', axis=1, inplace=True)
````

````python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from collections import Counter

# Podziel dane na testowe i treningowe
X_train, X_test, y_train, y_test = train_test_split(df_labeled.drop('popularity_labels_encoded', axis=1), df_labeled['popularity_labels_encoded'], test_size=0.2, random_state=42)

# smote = SMOTE(random_state=42)
# X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# print(f"Przed oversamplingiem: {Counter(y_train)}")
# print(f"Po oversamplingu: {Counter(y_train_resampled)}")
````

````python
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import History
from keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam


X_train = X_train.values
y_train = y_train.values
X_test = X_test.values
y_test = y_test.values

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = Sequential()
model.add(Dense(32, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(5, activation='softmax'))

history = History()

model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])

#early_stop = EarlyStopping(monitor='val_loss', patience=8)
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test),callbacks=[history])

y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
train_loss, train_acc = model.evaluate(X_train, y_train)
test_loss, test_acc = model.evaluate(X_test, y_test)

print('Strata na danych treningowych:', train_loss)
print('Dok≈Çadno≈õƒá na danych treningowych:', train_acc)
print('Strata na danych testowych:', test_loss)
print('Dok≈Çadno≈õƒá na danych testowych:', test_acc)

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Values')
plt.legend()
plt.show()
````

````python
from sklearn.metrics import confusion_matrix

cm_model = confusion_matrix(y_test, y_pred_classes)
print('\nMacierz pomy≈Çek dla sieci neuronowej:')
print(cm_model)
sns.heatmap(cm_model, annot=True, fmt='d', cmap='YlGnBu')

# Add labels and title
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Decision Tree Confusion Matrix')
````
